```python
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "text.latex.preamble": r"\usepackage[utf8]{inputenc}\usepackage[T1]{fontenc}\usepackage{lmodern}"
})
```

```python
# Parametry
s = 100  # Parametr do dostosowania
# Wartości osi X
d = np.arange(1, 1001, 1)

# Funkcje
n1 = s**2 * np.log(d)
n2 = s * np.log(np.e * d / s)

plt.figure(figsize=(6, 4))

# Dodanie informacji o s w legendzie
plt.plot(d, n1, label=fr'$n = s^2 \log(d),\ s={s}$', color='red', markersize=7, linewidth=1)
plt.plot(d, n2, label=fr'$n = s \log\left(\frac{{ed}}{{s}}\right),\ s={s}$', color='blue', markersize=7, linewidth=1)

plt.xlabel(r'Wymiar $d$', fontsize=13)
plt.ylabel(r'Wymagana wielkość próby $n$', fontsize=13)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=13)

os.makedirs("zdjecia_symulacji", exist_ok=True)
plt.savefig("zdjecia_symulacji/funkcje_n_od_d.svg", format="svg", bbox_inches='tight', pad_inches=0)

plt.show()
```
