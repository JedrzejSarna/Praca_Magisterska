```python
import numpy as np
import matplotlib.pyplot as plt
import os
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "text.latex.preamble": r"\usepackage{amsmath,amsfonts,amssymb}"
})
```

```python
# Parametry theta, t i współrzędne Δ
theta1, theta2 = 0, 2
t = 1  
Delta_x, Delta_y = 1, -2  # Współrzędne Δ
Delta = (Delta_x, Delta_y)

# Punkt theta + Δ * t
theta_plus_Delta_t = (theta1 + Delta_x * t, theta2 + Delta_y * t)

# Obliczenia dla kuli l1
suma_theta = np.abs(theta1) + np.abs(theta2)  # |\theta_1| + |\theta_2|
romb_x_1 = [suma_theta, 0, -suma_theta, 0]
romb_y_1 = [0, suma_theta, 0, -suma_theta]

# Obliczenia dla stożka stycznego przesuniętego
romb_x_2 = [
    (suma_theta - theta1) / t,
    -theta1 / t,
    -(suma_theta + theta1) / t,
    -theta1 / t
]
romb_y_2 = [
    -theta2 / t,
    (suma_theta - theta2) / t,
    -theta2 / t,
    -(suma_theta + theta2) / t
]

# Rysowanie wykresu
plt.figure(figsize=(6, 6))

# kula l1
plt.fill(romb_x_1, romb_y_1, color='lightblue', alpha=0.5, label=r"$\mathbb{B}_1(\| \theta^* \|_1)$")

# Stozek styczny
plt.fill(romb_x_2, romb_y_2, color='darkseagreen', alpha=0.4, label=r"$\mathbb{T}(\theta^*)$")

# Punkty: theta^*, Δ, i theta^* + Δ * t
plt.scatter([theta1], [theta2], color='royalblue', zorder=5, label=r"$\theta^*$")
plt.scatter([Delta_x], [Delta_y], color='darkgreen', zorder=5, label=r"$\Delta$")
plt.scatter([theta_plus_Delta_t[0]], [theta_plus_Delta_t[1]], color='darkorange', zorder=5, label=r"$\theta^* + t \Delta$")

# Dostosowanie wykresu
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel(r'$x$', fontsize=16)
plt.ylabel(r'$y$', fontsize=16)
plt.legend(fontsize=16)

os.makedirs("zdjecia_symulacji", exist_ok=True)
plt.savefig("zdjecia_symulacji/stozek_styczny_1.svg", format="svg")

plt.show()
```

```python
# Parametry theta i t
theta1, theta2 = 0, 2  # Współrzędne theta^*
t = 1  # Wartość t

# Obliczenia dla stożka stycznego przesuniętego
suma_theta = np.abs(theta1) + np.abs(theta2)  # |\theta_1| + |\theta_2|
romb_x_2 = [
    (suma_theta - theta1) / t + theta1,  # Wierzchołek przesunięty w prawo
    -theta1 / t + theta1,                # Wierzchołek przesunięty w górę
    -(suma_theta + theta1) / t + theta1, # Wierzchołek przesunięty w lewo
    -theta1 / t + theta1                 # Wierzchołek przesunięty w dół
]
romb_y_2 = [
    -theta2 / t + theta2,                # Wierzchołek przesunięty w prawo
    (suma_theta - theta2) / t + theta2,  # Wierzchołek przesunięty w górę
    -theta2 / t + theta2,                # Wierzchołek przesunięty w lewo
    -(suma_theta + theta2) / t + theta2  # Wierzchołek przesunięty w dół
]

# Definicja prostej \(y = ax + \theta_2\)
a = 0.5
x_vals = np.linspace(-5, 5, 100)
y_line = a * x_vals + theta2  # y = ax + theta2

# Rysowanie wykresu
plt.figure(figsize=(6, 6))

# Zielony obszar (stożek styczny przesunięty)
plt.fill(romb_x_2, romb_y_2, color='darkseagreen', alpha=0.4, label=r"$\theta^* + \mathbb{T}(\theta^*)$")

# Czerwona linia (jadro macierzy X)
plt.plot(x_vals, y_line, color='brown', label=r"$\theta^* + \text{ker}(\mathbf{X})$", linewidth=1.5)

# Punkt \(\theta^*\)
plt.scatter([theta1], [theta2], color='royalblue', label=r"$\theta^*$", zorder=5)

# Dostosowanie wykresu
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel(r"$x$", fontsize=16)
plt.ylabel(r"$y$", fontsize=16)
plt.legend(fontsize=16, loc="upper left")

plt.savefig("zdjecia_symulacji/stozek_styczny_przesuniety_1.svg", format="svg")

plt.show()
```
