```python
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "text.latex.preamble": r"\usepackage[utf8]{inputenc}\usepackage[T1]{fontenc}\usepackage{lmodern}"
})
```

```python
C = 1
d = 10  # Rozmiar wektora
j = np.arange(1, d + 1, dtype=float)

# Rne wartoci alpha
alpha_1 = 1
alpha_2 = 2
alpha_5 = 5

# Wartoci wektora dla rnych alpha
theta_alpha1 = C * j ** -alpha_1
theta_alpha2 = C * j ** -alpha_2
theta_alpha5 = C * j ** -alpha_5

# Rysowanie wykresw
plt.figure(figsize=(10, 5))

plt.plot(j, theta_alpha1, label=r'$\alpha = 1$', marker='s', color='gray', markersize=9, linewidth=1)  # Thinner line
plt.plot(j, theta_alpha2, label=r'$\alpha = 2$', marker='o', color='gray', markersize=9, linewidth=1)  # Thinner line
plt.plot(j, theta_alpha5, label=r'$\alpha = 5$', marker='^', color='gray', markersize=9, linewidth=1)  # Thinner line

# O X i Y
plt.xlabel(r'Indeks $j$', fontsize=20)
plt.ylabel(r'Warto ograniczenia $Cj^{-\alpha}$', fontsize=20)

# Dodanie kadego punktu na osi X
plt.xticks(j, fontsize=16) 
plt.yticks(fontsize=16)

plt.legend(fontsize=20)

os.makedirs("zdjecia_symulacji", exist_ok=True)
plt.savefig("zdjecia_symulacji/wektor_kompresowalny_1.svg", format="svg", bbox_inches='tight', pad_inches=0)

plt.show()
```

```python
C = 1
d = 10  # Rozmiar wektora
j = np.arange(1, d + 1, dtype=float)

# Rne wartoci alpha
alpha_1 = 5
alpha_2 = 7
alpha_5 = 10

# Wartoci wektora dla rnych alpha
theta_alpha1 = C * j ** -alpha_1
theta_alpha2 = C * j ** -alpha_2
theta_alpha5 = C * j ** -alpha_5

plt.figure(figsize=(10, 6))

# Dodanie wikszych punktw z ciesz lini czc
plt.plot(j, theta_alpha1, label=r'$\alpha = 5$', marker='s', color='gray', markersize=6, linewidth=1)  # Thinner line
plt.plot(j, theta_alpha2, label=r'$\alpha = 7$', marker='o', color='gray', markersize=6, linewidth=1)  # Thinner line
plt.plot(j, theta_alpha5, label=r'$\alpha = 10$', marker='^', color='gray', markersize=6, linewidth=1)  # Thinner line

# O X i Y
plt.xlabel(r'Indeks $j$', fontsize=20)
plt.ylabel(r'Warto ograniczenia $Cj^{-\alpha}$', fontsize=20)

# Dodanie kadego punktu na osi X
plt.xticks(j, fontsize=16)  
plt.yticks(fontsize=16)

plt.legend(fontsize=20)
plt.savefig("zdjecia_symulacji/wektor_kompresowalny_2.svg", format="svg", bbox_inches='tight', pad_inches=0)

# Wywietlenie wykresu
plt.show()
```
